````python
# Checking Null Values
print(sm.isnull().sum())
print('-*-' * 20)
print(teen.isnull().sum())
````

````python
# Checking Duplicate Values
print(sm.duplicated().sum())
print(teen.duplicated().sum())
# If duplicate present
# sm.drop_duplicates(inplace=True)
# teen.drop_duplicates(inplace=True)
````

````python
# Fixing Column Names
sm.columns = sm.columns.str.strip().str.lower().str.replace(" ", "_")
teen.columns = teen.columns.str.strip().str.lower().str.replace(" ", "_")
````

````python
# Fixing Datatypes
sm['age'] = sm['age'].astype(int)
sm['avg_daily_usage_hours'] = sm['avg_daily_usage_hours'].astype(int)
````

````python
# Fixing Datatypes
teen['age'] = teen['age'].astype(int)
teen['daily_social_media_hours'] = teen['daily_social_media_hours'].astype(int)
````

````python
# Fill Numeric nulls with median
# Why Median not mean?
# Median is robust to outliers
# Ex:
# values = [10, 20, 30, 1000]
# mean = 265
# median = 25    # better
````

````python
# Numeric columns
for col in sm.select_dtypes('number').columns:
    sm[col] = sm[col].fillna(sm[col].median())

for col in teen.select_dtypes('number').columns:
    teen[col] = teen[col].fillna(teen[col].median())
````

````python
# Categorical columns
for col in sm.select_dtypes(include=['object', 'string']):
    sm[col] = sm[col].fillna(sm[col].mode()[0])

for col in teen.select_dtypes(include=['object', 'string']):
    teen[col] = teen[col].fillna(teen[col].mode()[0])

# for col in sm.select_dtypes(include='object'):
#     sm[col] = sm[col].fillna(sm[col].mode()[0])

# for col in teen.select_dtypes(include='object'):
#     teen[col] = teen[col].fillna(teen[col].mode()[0])
````

## Encode Categorical Columns

````python
sm.columns
````

````python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cat_cols_sm = ['gender', 'academic_level', 'country', 'most_used_platform',
            'affects_academic_performance', 'overall_impact']
for col in cat_cols_sm:
    sm[col] = le.fit_transform(sm[col])
````

````python
teen.columns
````

````python
cat_cols_teen = ['gender', 'platform_usage', 'social_interaction_level']
for col in cat_cols_teen:
    teen[col] = le.fit_transform(teen[col])
````

````python
sm.head()
````

````python
teen.head()
````

````python
# Check Outliers (IQR)
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    return df[(df[col] >= Q1 - 1.5*IQR) & (df[col] <= Q3 + 1.5*IQR)]

sm = remove_outliers(sm, 'avg_daily_usage_hours')
teen = remove_outliers(teen, 'daily_social_media_hours')

print(sm.shape, teen.shape)
````

````python
sm.to_csv("../data/cleaned/social_media_cleaned.csv", index=False)
teen.to_csv("../data/cleaned/teen_mental_health_cleaned.csv", index=False)

print("Cleaned data saved!")
````
